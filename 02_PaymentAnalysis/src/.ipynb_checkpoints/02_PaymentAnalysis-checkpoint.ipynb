**Importing libraries (Numpy, Pandas, Geopandas, Shapely, Pysal, Matplotlib)**

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import pysal as ps
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from shapely.geometry import Point
%matplotlib inline
```

**Setting up global paths and file names**

```python
inputDataPath = '../input/'
outputDataPath = '../output/'
yellowFileName = 'SelectedYellowTrips_181015_181021.csv'
greenFileName = 'SelectedGreenTrips_181015_181021.csv'
taxiZonesFileName = 'taxi_zones.shp'
```

**Loading taxi data**

```python
yellowData = pd.read_csv(inputDataPath + yellowFileName, delimiter=',', index_col=0)
greenData = pd.read_csv(inputDataPath + greenFileName, delimiter=',', index_col=0)
```

**Loading taxi zones data**

```python
taxiZones = gpd.read_file(inputDataPath + taxiZonesFileName)
```

**Aggregating taxi data by zone and payment type** (1 = Credit Card; 2 = Cash)

```python
yellowCash = yellowData[yellowData['payment_type'] == 2]
yellowCredit = yellowData[yellowData['payment_type'] == 1]
greenCash = greenData[greenData['payment_type'] == 2]
greenCredit = greenData[greenData['payment_type'] == 1]
```

```python
yellowCash.shape
```

```python
yellowCredit.shape
```

```python
greenCash.shape
```

```python
greenCredit.shape
```

```python
yellowCashOriginsCount = yellowCash.groupby('PULocationID').count()
yellowCreditOriginsCount = yellowCredit.groupby('PULocationID').count()
greenCashOriginsCount = greenCash.groupby('PULocationID').count()
greenCreditOriginsCount = greenCredit.groupby('PULocationID').count()
```

```python
yellowCashDestinationCount = yellowCash.groupby('DOLocationID').count()
yellowCreditDestinationCount = yellowCredit.groupby('DOLocationID').count()
greenCashDestinationCount = greenCash.groupby('DOLocationID').count()
greenCreditDestinationCount = greenCredit.groupby('DOLocationID').count()
```

```python
yellowCashOriginsCount = yellowCashOriginsCount[['VendorID']]
yellowCreditOriginsCount = yellowCreditOriginsCount[['VendorID']]
greenCashOriginsCount = greenCashOriginsCount[['VendorID']]
greenCreditOriginsCount = greenCreditOriginsCount[['VendorID']]
yellowCashDestinationCount = yellowCashDestinationCount[['VendorID']]
yellowCreditDestinationCount = yellowCreditDestinationCount[['VendorID']]
greenCashDestinationCount = greenCashDestinationCount[['VendorID']]
greenCreditDestinationCount = greenCreditDestinationCount[['VendorID']]
```

```python
yellowCashOriginsCount.rename(columns={'VendorID':'YellowCashOriginCount'}, inplace=True)
yellowCreditOriginsCount.rename(columns={'VendorID':'YellowCreditOriginCount'}, inplace=True)
greenCashOriginsCount.rename(columns={'VendorID':'GreenCashOriginCount'}, inplace=True)
greenCreditOriginsCount.rename(columns={'VendorID':'GreenCreditOriginCount'}, inplace=True)
yellowCashDestinationCount.rename(columns={'VendorID':'YellowCashDestinationCount'}, inplace=True)
yellowCreditDestinationCount.rename(columns={'VendorID':'YellowCreditDestinationCount'}, inplace=True)
greenCashDestinationCount.rename(columns={'VendorID':'GreenCashDestinationCount'}, inplace=True)
greenCreditDestinationCount.rename(columns={'VendorID':'GreenCreditDestinationCount'}, inplace=True)
```

```python
yellowOriginsCount = pd.merge(yellowCashOriginsCount, yellowCreditOriginsCount, how='outer', left_index=True, right_index=True)
yellowDestinationsCount = pd.merge(yellowCashDestinationCount, yellowCreditDestinationCount, how='outer', left_index=True, right_index=True)
greenOriginsCount = pd.merge(greenCashOriginsCount, greenCreditOriginsCount, how='outer', left_index=True, right_index=True)
greenDestinationsCount = pd.merge(greenCashDestinationCount, greenCreditDestinationCount, how='outer', left_index=True, right_index=True)
```

```python
yellowOriginsCount.head()
```

```python
yellowDestinationsCount.head()
```

```python
greenOriginsCount.head()
```

```python
greenDestinationsCount.head()
```

```python
yellowCounts = pd.merge(yellowOriginsCount, yellowDestinationsCount, how='outer', left_index=True, right_index=True)
greenCounts = pd.merge(greenOriginsCount, greenDestinationsCount, how='outer', left_index=True, right_index=True)
```

```python
yellowCounts.head()
```

```python
greenCounts.head()
```

```python
taxiCounts = pd.merge(yellowCounts, greenCounts, how='outer', left_index=True, right_index=True)
```

```python
taxiCounts.head()
```

```python
taxiCounts.sum()
```

**Joining taxi data to zones**

```python
paymentData = pd.merge(taxiZones, taxiCounts, how='left', left_on='LocationID', right_index=True)
```

```python
paymentData.head()
```

```python
paymentData.fillna(0, inplace=True)
```

**LISA (Local Spatial Autocorrelation)**

Spatial Weights Matrix

```python
paymentData['YellowOriginPercCash'] = paymentData['YellowCashOriginCount'] / (paymentData['YellowCashOriginCount'] + paymentData['YellowCreditOriginCount']) * 100
paymentData['YellowDestinationPercCash'] = paymentData['YellowCashDestinationCount'] / (paymentData['YellowCashDestinationCount'] + paymentData['YellowCreditDestinationCount']) * 100
paymentData['GreenOriginPercCash'] = paymentData['GreenCashOriginCount'] / (paymentData['GreenCashOriginCount'] + paymentData['GreenCreditOriginCount']) * 100
paymentData['GreenDestinationPercCash'] = paymentData['GreenCashDestinationCount'] / (paymentData['GreenCashDestinationCount'] + paymentData['GreenCreditDestinationCount']) * 100
```

```python
bins = {'bins':[10, 20, 30, 40, 50], 'k':6}
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 20))
ax1 = paymentData.plot(ax=ax1, column='YellowOriginPercCash', legend=True, cmap='Blues', linewidth=0.1, edgecolor='0.8', scheme='user_defined', classification_kwds=bins)
ax2 = paymentData.plot(ax=ax2, column='YellowDestinationPercCash', legend=False, cmap='Blues', linewidth=0.1, edgecolor='0.8', scheme='user_defined', classification_kwds=bins)
ax1.axis('off')
ax2.axis('off')
ax1.title.set_text('Yellow Taxis Origins Percent Cash Payments')
ax2.title.set_text('Yellow Taxis Destinations Percent Cash Payments')
```

```python
paymentData.fillna(0, inplace=True)
```

```python
# Create the spatial weights matrix
w = ps.lib.weights.Queen.from_dataframe(paymentData)
```

```python
# Drop islands
paymentData = paymentData.drop(w.islands)
```

```python
# Recalculate the spatial weights matrix
w = ps.lib.weights.Queen.from_dataframe(paymentData)
```

```python
# Row standardize the matrix
w.transform = 'R'
```

```python
w[0]
```

```python
# LISA (Local indicators of spatial association)
# Break observations into significant or not
# Store the quadrant they belong to
lisa = ps.explore.esda.Moran_Local(paymentData['YellowOriginPercCash'], w)
paymentData['YellowOriginPercCash_sig'] = lisa.p_sim < 0.05
paymentData['YellowOriginPercCash_quad'] = lisa.q
lisa = ps.explore.esda.Moran_Local(paymentData['YellowDestinationPercCash'], w)
paymentData['YellowDestinationPercCash_sig'] = lisa.p_sim < 0.05
paymentData['YellowDestinationPercCash_quad'] = lisa.q
lisa = ps.explore.esda.Moran_Local(paymentData['GreenOriginPercCash'], w)
paymentData['GreenOriginPercCash_sig'] = lisa.p_sim < 0.05
paymentData['GreenOriginPercCash_quad'] = lisa.q
lisa = ps.explore.esda.Moran_Local(paymentData['GreenDestinationPercCash'], w)
paymentData['GreenDestinationPercCash_sig'] = lisa.p_sim < 0.05
paymentData['GreenDestinationPercCash_quad'] = lisa.q
```

```python
variable1 = 'YellowOriginPercCash'
variable2 = 'YellowDestinationPercCash'

f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
high_high = mpatches.Patch(color='#ff9a00', label='High-high cluster')
high_low = mpatches.Patch(color='#ffcc7f', label='High-low cluster')
low_low = mpatches.Patch(color='#738fbf', label='Low-low cluster')
low_high = mpatches.Patch(color='#99a7bf', label='Low-high cluster')
# Plot insignificant clusters
ns = paymentData.loc[paymentData[variable1 + '_sig']==False, 'geometry']
ns.plot(ax=ax1, color='#f7f7f7', label='test')
# Plot HH clusters
hh = paymentData.loc[(paymentData[variable1 + '_quad']==1) & (paymentData[variable1 + '_sig']==True), 'geometry']
hh.plot(ax=ax1, color='#ff9a00')
# Plot LL clusters
ll = paymentData.loc[(paymentData[variable1 + '_quad']==3) & (paymentData[variable1 + '_sig']==True), 'geometry']
ll.plot(ax=ax1, color='#738fbf')
# Plot LH clusters
lh = paymentData.loc[(paymentData[variable1 + '_quad']==2) & (paymentData[variable1 + '_sig']==True), 'geometry']
lh.plot(ax=ax1, color='#99a7bf')
# Plot HL clusters
hl = paymentData.loc[(paymentData[variable1 + '_quad']==4) & (paymentData[variable1 + '_sig']==True), 'geometry']
hl.plot(ax=ax1, color='#ffcc7f')
# Style and draw
ax1.title.set_text('LISA for yellow taxis origin cash payment')
ax1.set_axis_off()

# Plot insignificant clusters
ns = paymentData.loc[paymentData[variable2 + '_sig']==False, 'geometry',]
ns.plot(ax=ax2, color='#f7f7f7')
# Plot HH clusters
hh = paymentData.loc[(paymentData[variable2 + '_quad']==1) & (paymentData[variable2 + '_sig']==True), 'geometry']
hh.plot(ax=ax2, color='#ff9a00')
# Plot LL clusters
ll = paymentData.loc[(paymentData[variable2 + '_quad']==3) & (paymentData[variable2 + '_sig']==True), 'geometry']
ll.plot(ax=ax2, color='#738fbf')
# Plot LH clusters
lh = paymentData.loc[(paymentData[variable2 + '_quad']==2) & (paymentData[variable2 + '_sig']==True), 'geometry']
lh.plot(ax=ax2, color='#99a7bf')
# Plot HL clusters
hl = paymentData.loc[(paymentData[variable2 + '_quad']==4) & (paymentData[variable2 + '_sig']==True), 'geometry']
hl.plot(ax=ax2, color='#ffcc7f')
# Style and draw
ax2.title.set_text('LISA for yellow taxis destination cash payment')
ax2.set_axis_off()
ax2.legend(loc='upper left', frameon=False, handles=[high_high, high_low, low_high, low_low])
```
